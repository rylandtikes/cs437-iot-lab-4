```python
import pandas as pd
import seaborn as sns
```

```python
v0 = pd.read_csv("vehicle0.csv")
v1 = pd.read_csv("vehicle1.csv")
v2 = pd.read_csv("vehicle2.csv")
v3 = pd.read_csv("vehicle3.csv")
v4 = pd.read_csv("vehicle4.csv")

# v0, v1, v2, v3, v4
```

```python
plot1 = sns.lmplot(data=v0, x="vehicle_speed", y="vehicle_CO2")
plot1
```

```python
vehicle0_CO2 = pd.Series(v0['vehicle_CO2'])
vehicle0_CO2.plot(kind='kde')
vehicle1_CO2 = pd.Series(v1['vehicle_CO2'])
vehicle1_CO2.plot(kind='kde')
vehicle2_CO2 = pd.Series(v2['vehicle_CO2'])
vehicle2_CO2.plot(kind='kde')
vehicle3_CO2 = pd.Series(v3['vehicle_CO2'])
vehicle3_CO2.plot(kind='kde')
vehicle4_CO2 = pd.Series(v4['vehicle_CO2'])
vehicle4_CO2.plot(kind='kde')
```
